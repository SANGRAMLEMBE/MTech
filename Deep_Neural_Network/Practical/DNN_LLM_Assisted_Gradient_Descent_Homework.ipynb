### What is Gradient Descent?

Gradient descent is an **optimization algorithm** used to minimize a function‚Äîin machine learning, commonly the cost (or loss) function.

- Think of the cost function as a hilly landscape where we want to find the lowest point (minimum cost).
- Gradient descent helps us **iteratively move downhill** on this cost landscape by taking small steps in the direction of the steepest descent.
- The "gradient" is a vector that points in the direction of the greatest increase of the function. So moving in the opposite direction of the gradient reduces the cost.

In mathematical terms, for a cost function \( J(\theta) \), where \(\theta\) represents the model parameters (weights and biases):

\[
\theta = \theta - \alpha \nabla_{\theta} J(\theta)
\]

- \(\nabla_{\theta} J(\theta)\) is the gradient of the cost with respect to \(\theta\).
- \(\alpha\) is the learning rate, controlling the step size.

---

### Why is gradient descent important?

- Many machine learning models, including deep neural networks, rely on gradient descent to **learn** the best parameters by minimizing the error on the training data.
- It is computationally efficient for large datasets and complex models.

---

### In linear regression:

- The cost function typically used is Mean Squared Error (MSE).
- Gradient descent updates weights and bias based on the derivative (gradient) of MSE to reach the best fit line minimizing the prediction error.


````python
# ü§ñ PASTE LLM RESPONSE HERE - Exercise 1
# Copy the gradient descent function from your LLM response below:

# YOUR LLM-GENERATED CODE GOES HERE

import numpy as np
import matplotlib.pyplot as plt

def simple_gradient_descent(x, y, learning_rate=0.01, epochs=1000):
    """
    Perform basic gradient descent for linear regression

    Parameters:
    x : numpy array of shape (N,) -- input features
    y : numpy array of shape (N,) -- target values
    learning_rate : float -- step size for gradient descent updates
    epochs : int -- number of iterations to run gradient descent

    Returns:
    w : float -- final weight after training
    b : float -- final bias after training
    cost_history : list -- list of cost values at each epoch
    """

    # Number of training examples
    N = len(y)

    # Initialize parameters: weight and bias to zero
    w = 0.0
    b = 0.0

    cost_history = []

    for epoch in range(epochs):
        # Predicted output using current weights and bias
        y_pred = w * x + b

        # Calculate the error/residuals between predictions and true values
        error = y_pred - y

        # Mean Squared Error cost function:
        # J = (1/N) * sum(error^2)
        cost = (1 / N) * np.sum(error ** 2)
        cost_history.append(cost)

        # Derivatives of cost J with respect to w and b:
        # Derivative w.r.t w:
        # dJ/dw = (2/N) * sum(error * x)
        dw = (2 / N) * np.dot(error, x)

        # Derivative w.r.t b:
        # dJ/db = (2/N) * sum(error)
        db = (2 / N) * np.sum(error)

        # Update parameters by moving against the gradient direction
        w -= learning_rate * dw
        b -= learning_rate * db

        # Print progress every 100 epochs
        if (epoch + 1) % 100 == 0:
            print(f"Epoch {epoch+1}/{epochs}: Cost={cost:.4f}, w={w:.4f}, b={b:.4f}")

    return w, b, cost_history



````

# ü§ñ PASTE LLM RESPONSE HERE - Exercise 2
# Copy both functions from your LLM response:


- **Batch Gradient Descent (BGD):**  
  Evaluates the gradient using *all* samples in the dataset before each parameter update. This provides a stable and smooth descent path but can be slow per iteration for large datasets.

- **Stochastic Gradient Descent (SGD):**  
  Updates parameters for each individual sample (one at a time). It is noisy and can jump around during convergence but can start improving parameters faster because updates happen more frequently.

---

### Why:
- **SGD converges faster initially:**  
  Because it updates the model more frequently (per sample), it can quickly move towards a good solution even though the path is noisy.

- **BGD gives smoother convergence:**  
  Because gradients are averaged over the entire dataset, updates are more stable and less noisy, producing smooth cost function decrease.

---

### When to use:
- Use **BGD** when the dataset is small to medium and precise convergence path/stability is desired.
- Use **SGD** when the dataset is large; it allows faster approximate convergence and can better escape local minima in complex problems.

````python
# YOUR LLM-GENERATED CODE FOR BATCH GD:
import numpy as np
import time
import matplotlib.pyplot as plt

def batch_gradient_descent(x, y, learning_rate=0.01, epochs=1000):
    """
    Batch Gradient Descent:
    Use all samples per update to compute the gradient.

    Parameters:
    x -- numpy array of shape (N, d), N samples, d features
    y -- numpy array of shape (N,)
    learning_rate -- step size
    epochs -- iterations to run

    Returns:
    w -- weights numpy array shape (d,)
    b -- bias scalar
    cost_history -- list of cost values over epochs
    """
    N, d = x.shape
    w = np.zeros(d)
    b = 0.0
    cost_history = []

    start_time = time.time()

    for epoch in range(epochs):
        # Predictions for all samples
        y_pred = np.dot(x, w) + b

        # Compute error
        error = y_pred - y

        # Compute mean squared error cost
        cost = (1 / N) * np.sum(error ** 2)
        cost_history.append(cost)

        # Compute gradients
        dw = (2 / N) * np.dot(x.T, error)  # shape (d,)
        db = (2 / N) * np.sum(error)

        # Parameter update
        w -= learning_rate * dw
        b -= learning_rate * db

        # Print progress every 100 epochs
        if (epoch + 1) % 100 == 0:
            print(f"[BGD] Epoch {epoch + 1}/{epochs}: Cost={cost:.4f}")

    end_time = time.time()
    print(f"Batch GD took {end_time - start_time:.4f} seconds")

    return w, b, cost_history


# YOUR LLM-GENERATED CODE FOR STOCHASTIC GD:
def stochastic_gradient_descent(x, y, learning_rate=0.01, epochs=1000):
    """
    Stochastic Gradient Descent:
    Use one sample at a time to compute and update gradients.
    Shuffle dataset each epoch.

    Parameters:
    x -- numpy array of shape (N, d)
    y -- numpy array of shape (N,)

    Returns:
    w -- weights numpy array shape (d,)
    b -- bias scalar
    cost_history -- list of mean cost computed at end of each epoch
    """
    N, d = x.shape
    w = np.zeros(d)
    b = 0.0
    cost_history = []

    start_time = time.time()

    for epoch in range(epochs):
        # Shuffle the data indices for each epoch
        indices = np.arange(N)
        np.random.shuffle(indices)

        for i in indices:
            xi = x[i,:]
            yi = y[i]

            # Prediction for single sample
            y_pred = np.dot(xi, w) + b

            # Error for this sample
            error = y_pred - yi

            # Gradients (for one sample)
            dw = 2 * error * xi
            db = 2 * error

            # Update parameters
            w -= learning_rate * dw
            b -= learning_rate * db

        # Compute cost after epoch (over full dataset for tracking)
        y_pred_all = np.dot(x, w) + b
        cost = (1 / N) * np.sum((y_pred_all - y) ** 2)
        cost_history.append(cost)

        if (epoch + 1) % 100 == 0:
            print(f"[SGD] Epoch {epoch + 1}/{epochs}: Cost={cost:.4f}")

    end_time = time.time()
    print(f"Stochastic GD took {end_time - start_time:.4f} seconds")

    return w, b, cost_history


# Any test/comparison code from LLM:
# Comparison test with synthetic multi-feature data:
np.random.seed(42)
N = 200
d = 3

# Synthetic features randomly between 0 and 10
x = np.random.rand(N, d) * 10

# True weights and bias
true_w = np.array([1.5, -2.0, 3.0])
true_b = 5.0

# Targets with noise
noise = np.random.normal(0, 1, N)
y = np.dot(x, true_w) + true_b + noise

# Run both methods
w_bgd, b_bgd, cost_bgd = batch_gradient_descent(x, y, learning_rate=0.001, epochs=1000)
w_sgd, b_sgd, cost_sgd = stochastic_gradient_descent(x, y, learning_rate=0.001, epochs=1000)

# Plot cost histories
plt.figure(figsize=(10,6))
plt.plot(cost_bgd, label='Batch Gradient Descent')
plt.plot(cost_sgd, label='Stochastic Gradient Descent')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error Cost')
plt.title('BGD vs SGD Convergence')
plt.legend()
plt.grid(True)
plt.show()

print(f"BGD final weights: {w_bgd}, bias: {b_bgd:.4f}")
print(f"SGD final weights: {w_sgd}, bias: {b_sgd:.4f}")
````

````python
# ü§ñ PASTE LLM RESPONSE HERE - Exercise 3
# Copy the mini-batch GD function and comparison code:
import numpy as np
import time
import matplotlib.pyplot as plt

def mini_batch_gradient_descent(x, y, batch_size=32, learning_rate=0.01, epochs=1000):
    """
    Mini-batch gradient descent for linear regression.

    Parameters:
    x -- numpy array shape (N, d), features
    y -- numpy array shape (N,), targets
    batch_size -- int, number of samples per batch
    learning_rate -- float, step size
    epochs -- int, number of epochs

    Returns:
    w -- numpy array (d,), weights
    b -- float, bias
    cost_history -- list of cost at end of each epoch
    batches_per_epoch -- int, number of batches processed per epoch
    """
    N, d = x.shape
    w = np.zeros(d)
    b = 0.0
    cost_history = []

    start_time = time.time()

    # Calculate number of batches per epoch accounting for any leftover samples
    batches_per_epoch = int(np.ceil(N / batch_size))

    for epoch in range(epochs):
        # Shuffle the dataset indices
        indices = np.arange(N)
        np.random.shuffle(indices)

        # Process mini-batches
        for batch_num in range(batches_per_epoch):
            start_idx = batch_num * batch_size
            end_idx = min(start_idx + batch_size, N)
            batch_indices = indices[start_idx:end_idx]

            x_batch = x[batch_indices]
            y_batch = y[batch_indices]

            # Prediction for batch
            y_pred = np.dot(x_batch, w) + b

            # Error
            error = y_pred - y_batch

            # Gradients
            dw = (2 / len(y_batch)) * np.dot(x_batch.T, error)
            db = (2 / len(y_batch)) * np.sum(error)

            # Parameter updates
            w -= learning_rate * dw
            b -= learning_rate * db

        # Compute cost over full dataset after epoch
        y_pred_all = np.dot(x, w) + b
        cost = (1 / N) * np.sum((y_pred_all - y) ** 2)
        cost_history.append(cost)

        if (epoch + 1) % 100 == 0:
            print(f"[Mini-batch GD] Epoch {epoch + 1}/{epochs}: Cost={cost:.4f} Batch size={batch_size}")

    end_time = time.time()
    print(f"Mini-batch GD with batch size {batch_size} took {end_time - start_time:.4f} seconds")

    return w, b, cost_history, batches_per_epoch

# YOUR LLM-GENERATED MINI-BATCH GD FUNCTION:


# YOUR LLM-GENERATED BATCH SIZE COMPARISON CODE:
def compare_batch_sizes(x, y, batch_sizes, learning_rate=0.01, epochs=500):
    """
    Compare convergence patterns of different batch sizes including 'full' batch.

    Parameters:
    x -- numpy array features
    y -- numpy array targets
    batch_sizes -- list of batch sizes, 'full' string supported for full batch gradient descent
    learning_rate -- float
    epochs -- int
    """
    plt.figure(figsize=(12, 7))

    for bs in batch_sizes:
        if bs == 'full':
            print(f"\nRunning full Batch Gradient Descent...")
            w, b, cost_hist = batch_gradient_descent(x, y, learning_rate=learning_rate, epochs=epochs)
            label = 'Full Batch Gradient Descent'
            batches_per_epoch = 1
        else:
            print(f"\nRunning Mini-batch Gradient Descent with batch size = {bs}...")
            w, b, cost_hist, batches_per_epoch = mini_batch_gradient_descent(x, y,
                                                                            batch_size=bs,
                                                                            learning_rate=learning_rate,
                                                                            epochs=epochs)
            label = f"Mini-batch size {bs}"

        plt.plot(cost_hist, label=f"{label} (batches/epoch={batches_per_epoch})")

    plt.xlabel('Epochs')
    plt.ylabel('Mean Squared Error Cost')
    plt.title(f'Convergence patterns for different batch sizes (lr={learning_rate})')
    plt.legend()
    plt.grid(True)
    plt.show()
````
