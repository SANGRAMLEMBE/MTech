```python
# %% [markdown]
# # Role of Statistics in Data Science
#
# Statistics is the backbone of Data Science.
# It helps us:
# - Summarize and interpret data
# - Understand patterns and relationships
# - Make predictions and decisions
# - Validate assumptions using hypothesis testing
#
# **In this notebook, we will cover:**
# 1. Types of Data
# 2. Random Variables (Numerical & Categorical)
# 3. Data Collection & Sampling Techniques
# 4. Descriptive Statistics
# 5. Data Visualization

# %% [markdown]
# ## 1. Types of Data
#
# Data can be broadly categorized into:
# - **Numerical (Quantitative)**: Numbers representing quantities (e.g., height, salary, marks)
# - **Categorical (Qualitative)**: Labels or categories (e.g., gender, color, city)

# %%
# Numerical Data
numerical_data = [12, 15, 18, 20, 21, 22, 25, 30]

# Categorical Data
categorical_data = ["Red", "Blue", "Red", "Green", "Blue", "Red"]

print("Numerical Data:", numerical_data)
print("Categorical Data:", categorical_data)

# %% [markdown]
# ## 2. Random Variables
#
# - **Numerical Random Variable**: Takes numerical values based on a probability distribution.
# - **Categorical Random Variable**: Takes category labels with certain probabilities.

# %%
import numpy as np

np.random.seed(0)

# Numerical Random Variable (Normal Distribution)
normal_rv = np.random.normal(loc=50, scale=10, size=100)
print("Numerical Random Variable Sample (first 5):", normal_rv[:5])

# Categorical Random Variable
categories = ["Yes", "No"]
categorical_rv = np.random.choice(categories, size=20, p=[0.6, 0.4])
print("Categorical Random Variable:", categorical_rv)

# %% [markdown]
# ## 3. Data Collection & Sampling
#
# Data can be collected through:
# - Surveys, experiments, sensors, web scraping, etc.
#
# Sampling methods:
# - **Random Sampling**: Each member has an equal chance.
# - **Systematic Sampling**: Select every *k*-th element.
# - **Stratified Sampling**: Split into groups and sample proportionally.

# %%
import pandas as pd

# Creating a sample dataset
data = pd.DataFrame({
    "ID": range(1, 21),
    "Age": np.random.randint(18, 60, size=20),
    "Gender": np.random.choice(["Male", "Female"], size=20),
    "Score": np.random.randint(40, 100, size=20)
})

print("Original Data:\n", data)

# Random Sampling
sample_random = data.sample(n=5, random_state=1)
print("\nRandom Sample:\n", sample_random)

# Systematic Sampling
step = 3
systematic_sample = data.iloc[::step, :]
print("\nSystematic Sample:\n", systematic_sample)

# Stratified Sampling (by Gender)
stratified_sample = data.groupby("Gender", group_keys=False).apply(lambda x: x.sample(2, random_state=1))
print("\nStratified Sample:\n", stratified_sample)

# %% [markdown]
# ## 4. Descriptive Statistics
#
# Descriptive statistics summarize the central tendency, dispersion, and shape of a dataset's distribution.
# Common measures:
# - Mean
# - Median
# - Mode
# - Standard Deviation

# %%
from scipy import stats

print("Descriptive Statistics for Score:\n", data["Score"].describe())

mean_score = np.mean(data["Score"])
median_score = np.median(data["Score"])
mode_score = stats.mode(data["Score"], keepdims=True)[0][0]
std_dev = np.std(data["Score"], ddof=1)

print(f"\nMean: {mean_score:.2f}")
print(f"Median: {median_score}")
print(f"Mode: {mode_score}")
print(f"Standard Deviation: {std_dev:.2f}")

# %% [markdown]
# ## 5. Data Visualization
#
# Visualizations help in understanding the data distribution and frequency.

# %%
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))

# Histogram for numerical random variable
plt.subplot(1, 2, 1)
sns.histplot(normal_rv, kde=True)
plt.title("Histogram of Numerical Random Variable")

# Count plot for categorical random variable
plt.subplot(1, 2, 2)
sns.countplot(x=categorical_rv)
plt.title("Count of Categorical Random Variable")

plt.tight_layout()
plt.show()
```
