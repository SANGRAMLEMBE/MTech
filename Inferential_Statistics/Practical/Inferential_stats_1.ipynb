# Normal Random Variable ( Normal Distribution)

1. loc : float or array_like of floats
    Mean ("centre") of the distribution.
2. scale : float or array_like of floats
    Standard deviation (spread or "width") of the distribution. Must be non-negative.
3. size : int or tuple of ints, optional

```python
normal_random_variable = np.random.normal(loc = 40,scale= 10,size = 50 )
print(normal_random_variable)
```

```python
# Taking 1st five values of Random Variable
print("1st five Numerical Random Variable Sample: ",normal_random_variable[:5])
```

```python
arr = normal_random_variable[:5]
round_arr = np.round(arr,2)          # Value Round off till 2 decimal
print("1st five Numerical Random Variable Sample:",round_arr)
```

# Mean, Median, Mode

- Mean: The average of all values, calculated as the sum of all values divided by the number of values.

- Median: The middle value in a dataset when the data is arranged in order.

- Mode: The value that appears most frequently in a dataset

```python
print("Mean Age: ",Created_data['Age'].mean())      # mean  of Age
print("Median Age: ",Created_data['Age'].median())  # Median of Age
print("Mode Age: ",Created_data['Age'].mode()[0])   # Mode of Age

# in pandas can get more than 2 repited values, so it will show 1st value / [0]
```

```python
print("Mean Score:", Created_data['Score'].mean())     #Mean of score
print("Median Score:", Created_data['Score'].median()) # Median of Score
print("Mode Score:", Created_data['Score'].mode()[0])  # Mode of Score
```

# Variance and Standard Variance

- variance =  = (x - ) / N

- standard deviation =  = Square root of (variance)

```python
print("Age Variance:", Created_data['Age'].var())
print("Age Standard Deviation:", Created_data['Age'].std())

```

```python
print("Score Variance:", Created_data['Score'].var())
print("Score Standard Deviation:", Created_data['Score'].std())
```

#Skewness

- Positive skew (right-skewed): Tail is longer on the right side; mean > median.

- Negative skew (left-skewed): Tail is longer on the left side; mean < median.

- Zero skew: Symmetrical distribution where mean = median = mode.

```python
print("Age Skewness:", Created_data['Age'].skew())
print("Score Skewness:", Created_data['Score'].skew())

```

# Kurtosis

- High kurtosis (leptokurtic) means more extreme outliers and fat tails.

- Low kurtosis (platykurtic) means fewer outliers and thin tails.

- Medium kurtosis (mesokurtic) is typical of a normal distribution.

```python
print("Age Kurtosis:", Created_data['Age'].kurt())
print("Score Kurtosis:", Created_data['Score'].kurt())
```

# Random Sampling
- Every member of the population has an equal chance of being selected.
- Respects the principle of independence

```python
random_sample = Created_data.sample(n=10,random_state= 1)
print("Random_sample: \n",random_sample)                 # sample data
```

# Systematic Sampling

-  Selects every kth individual from a list or sequence.

```python
systematic_sampling = Created_data.iloc[:30:3]   # here we taking step = 3
print("Systematic Sampling \n",systematic_sampling)
```

# Stratified Sampling (by Gender)

- The population is divided into groups based on a characteristic (e.g., age,
gender), and a random sample is taken from each groups.

```python
stratified_sample = Created_data.groupby("Gender", group_keys=False).apply(lambda x: x.sample(2, random_state=1))
print("Stratified Sample:\n", stratified_sample)
```

# T2: Building programs to work with Descriptive statistics in Python


```python
#load the data set
housing=fetch_california_housing(as_frame=True)
df=housing.frame
```

```python
# Display first few rows
df.head()
```

```python
# Shape of dataset
print("Shape of dataset:", df.shape)
```

```python
# Descriptive statistics in detail
desc=df.describe().T
print(desc)
```

```python
# Adding more statistics
desc["median"]=df.median()
desc["mode"]=df.mode().iloc[0]
desc["var"]=df.var()
desc["std"]=df.std()
desc["skewness"]=df.skew()
desc["kurtosis"]=df.kurtosis()
desc["range"]=df.max()-df.min()
desc["IQR"]=df.quantile(0.75)-df.quantile(0.25)
desc
```


# T3: Building programs to perform Correlation analysis in Python

```python
# Checking missing values
df.isnull().sum()
```

```python
# Histograms of all columns
df.hist(bins=30, figsize=(15,10), edgecolor='black')
plt.suptitle("Histograms of Features")
plt.show()
```

```python

# Boxplots of all features
plt.figure(figsize=(15,10))
for i, col in enumerate(df.columns, 1):
 plt.subplot(3,3,i)
 sns.boxplot(y=df[col], color="lightblue")
 plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Covariance matrix
corr=df.cov()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Covariance Heatmap")
plt.show()
```
