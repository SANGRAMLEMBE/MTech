## T4: Probability Computation

```python


# 1. Basic Probability
def calculate_probability(favorable_outcomes, total_outcomes):
    """
    Calculates the probability of an event.
    """
    if total_outcomes == 0:
        return 0
    return favorable_outcomes / total_outcomes
```

### 1. Basic Probability

- A patient is randomly chosen from a group of 5 diabetic patients,9 non-diabetic patients with a heart condition, and 11 non-diabetic patients with a liver condition.

```python
total_patients = 5 + 9 + 11
diabetic_patients = 5

probability_diabetic = calculate_probability(diabetic_patients, total_patients)
print(f"The probability of selecting a diabetic patient is: {probability_diabetic:.2f}")
```

### 2. Permutations


- A website requires a 4-character password using digits 0-9 without repetition.


```python

def calculate_permutations(n, r):
    """
    Calculates the number of permutations (arrangements).
    """
    return math.factorial(n) // math.factorial(n - r)

n_digits = 10
r_password_length = 4
possible_passwords = calculate_permutations(n_digits, r_password_length)
print(f"The number of possible 4-digit passwords is: {possible_passwords}")
```

### 3. Combinations

- Selecting 3 books from a set of 5 distinct books

```python
def calculate_combinations(n, r):
    """
    Calculates the number of combinations (selections).
    """
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

n_books = 5
r_selection = 3
possible_combinations = calculate_combinations(n_books, r_selection)
print(f"The number of ways to choose 3 books from 5 is: {possible_combinations}")
```

##  T5: Building programs work with Bayes theorem applications in Python



    Calculates the posterior probability P(A|B) using Bayes' theorem.

    Args:
        p_a: Prior probability of event A.
        p_b_given_a: Probability of event B given that A is true.
        p_b_given_not_a: Probability of event B given that A is false.

    Returns:
        The posterior probability of A given B.


```python
# T5: Bayes' Theorem Application

def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):

    # Calculate the probability of not A
    p_not_a = 1 - p_a

    # Calculate the probability of B
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)

    # Calculate the posterior probability P(A|B)
    p_a_given_b = (p_b_given_a * p_a) / p_b

    return p_a_given_b

```

### Example: Medical Diagnosis
Let's say a certain disease is present in 1% of the population.
A test for this disease is 99% accurate for people who have the disease and has a 5% false positive rate.


```python

# P(A): Probability of having the disease
p_disease = 0.01

# P(B|A): Probability of a positive test given the person has the disease
p_positive_given_disease = 0.99

# P(B|not A): Probability of a positive test given the person does not have the disease
p_positive_given_no_disease = 0.05

# Calculate the probability of having the disease given a positive test result
posterior_probability = bayes_theorem(p_disease, p_positive_given_disease, p_positive_given_no_disease)
print(f"The probability of having the disease given a positive test is: {posterior_probability:.2f}")
```

##  T6: Building programs work with Plotting and analyzing the distribution in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python

n_trials = 10
p_success = 0.5
x_binom = np.arange(0, n_trials + 1)
binomial_pmf = binom.pmf(x_binom, n=n_trials, p=p_success)

plt.figure(figsize=(8, 5))
plt.bar(x_binom, binomial_pmf, color='skyblue')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
lambda_param = 5  # Average number of events
x_poisson = np.arange(0, 15)
poisson_pmf = poisson.pmf(x_poisson, mu=lambda_param)

plt.figure(figsize=(8, 5))
plt.bar(x_poisson, poisson_pmf, color='salmon')
plt.title('Poisson Distribution (λ=5)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python
mu = 0       # Mean
sigma = 1    # Standard deviation
x_norm = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
normal_pdf = norm.pdf(x_norm, loc=mu, scale=sigma)

plt.figure(figsize=(8, 5))
plt.plot(x_norm, normal_pdf, color='green')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
